## Dickey fuller test

```python
from statsmodels.tsa.stattools import adfuller

dftest=adfuller(y,autolag='AIC')

dfoutput=pd.Series(dftest[0:4],index=['Test Statistic','p value','lags used','NUMBER OF OBSERVATION USED'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)' %key]=value
    
print(dfoutput)
```

```python
y_logScale=np.log(y)
plt.plot(y_logScale)
```

```python
sales_diff = Sale.diff(periods=1)
# integrated of order 2, denoted by d (foddr diff), one of the parameter of ARIMA model

sales_diff = sales_diff[1:]
sales_diff.head()
```

```python
sales_diff.plot()
```

# Checking trend and seasonality 

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(sales_diff,freq=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(sales_diff, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()

decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
```

```python
sales_diff1 = sales_diff.iloc[:,0].values
```

# Doing again Dickey fuller test

```python
from statsmodels.tsa.stattools import adfuller

dftest=adfuller(sales_diff1,autolag='AIC')

dfoutput=pd.Series(dftest[0:4],index=['Test Statistic','p value','lags used','NUMBER OF OBSERVATION USED'])
for key,value in dftest[4].items():
    dfoutput['critical value (%s)' %key]=value
    
print(dfoutput)
```

### Hence our model is stationary

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(sales_diff, nlags=20)
lag_pacf = pacf(sales_diff, nlags=20, method='ols')

#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(sales_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(sales_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(sales_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(sales_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```
